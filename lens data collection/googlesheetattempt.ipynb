```python
import pandas as pd
import os
import sys
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
```

```python
maindf = pd.read_csv("synbiolabraw.csv") 
maindf.columns = maindf.iloc[0]
maindf = maindf.tail(-1)
maindf.reset_index(drop=True, inplace=True)
maindf.head()
```

```python
maindf.columns= maindf.columns.str.lower()
maindf.rename({'name of pi': 'name', "research area:" : "research_focus","pi contact email address:" : "email", "organism(s) and methods:" : "methods", "assumed language(s) based on pi, website, and location:" : "assumed_languages", "twitter:" : "twitter"}, axis = "columns", inplace=True)
maindf.head()
```

```python
namecolumn = maindf.get("name")
maindf["name"] = maindf["name"].astype('str')
names = []
for rows in maindf:
    if rows == "name":
        for i in range(len(maindf[rows])):
            names.append(maindf[rows][i])
print(names)
```
